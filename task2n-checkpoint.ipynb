```python
import pandas as pd

# Load the dataset
data = pd.read_csv('C:\\Users\\kokat\\Downloads\\task2\\Global_Superstore(CSV).csv')  # Replace with your actual dataset path

# View first few rows of the dataset
print(data.head())

# Check for missing values and data types
print(data.info())
```

```python
# Fill missing values with the mean for 'Sales'
data['Sales'] = data['Sales'].fillna(data['Sales'].mean())

# Fill missing values with the median for 'Profit'
data['Profit'] = data['Profit'].fillna(data['Profit'].median())

```

```python
# Remove duplicate rows
data.drop_duplicates(inplace=True)
```

```python
# Identify and remove outliers using IQR for 'Sales'
Q1 = data['Sales'].quantile(0.25)
Q3 = data['Sales'].quantile(0.75)
IQR = Q3 - Q1

# Filter data to remove outliers
data = data[(data['Sales'] >= (Q1 - 1.5 * IQR)) & (data['Sales'] <= (Q3 + 1.5 * IQR))]
```

```python
# Calculate basic statistics for numerical columns
print(data[['Sales', 'Profit']].describe())
```

```python
# Select only numeric columns
numeric_data = data.select_dtypes(include=['number'])

# Calculate correlations between numerical columns
correlation_matrix = numeric_data.corr()
print(correlation_matrix)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot histogram for 'Sales'
sns.histplot(data['Sales'], kde=True)
plt.title('Sales Distribution')
plt.show()
```

```python
# Plot boxplot for 'Sales'
sns.boxplot(x=data['Sales'])
plt.title('Sales Boxplot')
plt.show()
```

```python
# Plot heatmap for correlations
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Load the dataset for Sales Performance Analysis
sales_data = pd.read_csv('C:\\Users\\kokat\\Downloads\\task2\\Global_Superstore(CSV).csv')  # Replace with your actual file path

# View first few rows and dataset information
print(sales_data.head())
print(sales_data.info())
```

```python
# View the first few rows
print(sales_data.head())

# Check for missing values and data types
print(sales_data.info())
```

```python
# Remove duplicate rows
sales_data.drop_duplicates(inplace=True)
```

```python
# Fill missing values with mean for 'Sales'
sales_data['Sales'] = sales_data['Sales'].fillna(sales_data['Sales'].mean())

# Fill missing values with median for 'Profit'
sales_data['Profit'] = sales_data['Profit'].fillna(sales_data['Profit'].median())

```

```python
# Check the column names in the dataset
print(sales_data.columns)

```

```python
# Convert the correct date column to datetime format
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'])
```

```python
# Set the correct column as index and plot Sales over time
sales_data.set_index('Order Date', inplace=True)
sales_data['Sales'].plot(figsize=(12, 6), title='Sales Over Time')
plt.show()
```

```python
# Scatter plot for Profit vs. Discount
sns.scatterplot(x='Discount', y='Profit', data=sales_data)
plt.title('Profit vs. Discount')
plt.show()
```

```python
# Bar plot for Region-wise Sales
sns.barplot(x='Region', y='Sales', data=sales_data)
plt.title('Sales by Region')
plt.show()
```

```python
# Pie chart for Category-wise Sales
category_sales = sales_data.groupby('Category')['Sales'].sum()
category_sales.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Sales Distribution by Category')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Define features and target
X = sales_data[['Profit', 'Discount']]
y = sales_data['Sales']

# Initialize and fit the model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Evaluate the model performance
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R Score: {r2}')
```
